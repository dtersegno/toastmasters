```python
cal = pd.DataFrame(columns = speakers + leaders + auxiliary, index = range(num_days))
cal.index.rename('day', inplace=True)

# create dicts of the students for each day and for each job.
# allowed students by day
full_student_list = [
        char for char in students
]

student_buckets_by_day = {
    day:full_student_list
    for day in cal.index
}
#allowed students by role
student_buckets_by_role = {
    role:full_student_list
    for role in speakers + leaders + auxiliary
}

#make an rng
seed = 2022
rng = np.random.default_rng(seed)
counter_students_max = 1000
verbose = False

#for each day,
for day in cal.index:
    student_buckets_by_day[day] = full_student_list
    #say which day is being filled.
    print(f'filling schedule for day {day}.')
    #for each role,
    for role in cal.columns:
#         if len(student_buckets_by_role[role]) == 0:
#             student_buckets_by_role[role] = full_student_list
        counter_students = 0
        role_selected = False
        while counter_students < counter_students_max and not role_selected:
            counter_students += 1
            #place a random student from the intersection of the role's student list and the day's list in the slot
            allowed_students = np.intersect1d(student_buckets_by_day[day], student_buckets_by_role[role])
            if len(allowed_students) == 0:
                print(f'No more allowed students found for day {day}, role {role}.')
                break
            else: #pick out a student and assign it that role
                this_student = rng.choice(allowed_students)
                cal.loc[day, role] = this_student
            #perform column checks
            #the column with itself (repeated role)
            columns_ok = check_job_repeats(cal.loc[:day], [role])
            # ...multiple columns for speaking roles, ...
            if role == prepared_speakers[-1]:
                columns_ok = columns_ok and check_job_type_repeats(cal.loc[:day], prepared_speakers, day)
            elif role == impromptu_speakers[-1]:
                columns_ok = columns_ok and check_job_type_repeats(cal.loc[:day], impromptu_speakers, day)
            elif role == evaluators[-1]:
                columns_ok = columns_ok and check_job_type_repeats(cal.loc[:day], evaluators, day)
            # ...and leadership roles.
            elif role == leaders[-1]:
                columns_ok = columns_ok and check_job_type_repeats(cal.loc[:day], leaders, day) 
            if counter_students%(counter_students_max//10) == 0 and counter_students%(counter_students_max)//10 > 0 and verbose:
                print(f'Tried {counter_students} combinations for day {day}, {role}.')
            if counter_students == counter_students_max and counter_students%(counter_students_max)//10 > 0 and verbose:
                print(f'Reached {counter_students_max} combinations for day {day}, {role}!')
            #check that the student hasn't been placed yet today.
            row_ok = check_day_repeats(cal.loc[[day]])
            if columns_ok and row_ok:
                role_selected = True
                print(f'Decisively placed {this_student} as {role}.')
                #the student has been chosen for the day
                student_buckets_by_day[day].remove(this_student)
                #the student has been chosen for this role
                try:
                    if role in prepared_speakers:
                        for column in prepared_speakers:
                            student_buckets_by_role[column].remove(this_student)
                    elif role in impromptu_speakers:
                        for column in impromptu_speakers:
                            student_buckets_by_role[column].remove(this_student)
                    elif role in leaders:
                        for column in evaluators:
                            student_buckets_by_role[column].remove(this_student)
                    else:
                        student_buckets_by_role[role].remove(this_student)
                except:
                    print(f'Could not remove student {this_student} for role {role} from {student_buckets_by_role[role]}.')
                
        #remove the student from the day's list of choices
print('done!')
        
```
